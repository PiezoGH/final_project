# Задание 3

В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

Будем представлять что у нас есть два месяца, в первый провели событие А, во второй событие В.
Оценивать результаты будем по разнице в показателях этих месяцев.

### По первой части:

Начать надо со стандартных DAU/WAU/MAU*.

Если от месяца к месяцу эти показатели растут, то можно говорить, что в целом дела идут правильно, игра/продукт (ивент) привлекательный.

Есть и более интересные метрики:
- Если количество подаренных бонусов в виде скинов, айтемов, монет (любая выдача призов) растет (скорее надо рассматривать среднее количество на игрока), то значит игрокам нравится ивент, они играют в него много.
- Надо смотреть на среднее время в игре/ивенте. Растёт - игра/ивент заходит аудитории.
- Рост ретеншена тоже скажет об удачности ивента
- Где смотрят ретеншен, там смотрят и Churn Rate. 
- Стоит ввести в игру/ивент ранги за успехи в ивенте, условные Gold/Silver/Bronze. Если соотношение растет (становится больше Gold, чем Silver/Bronze), то можно будет сказать, что игрокам нравится ивент.

### По второй части:

К набору метрик из первой части можно добавить подсчет количества откатов, смотреть среднее и абсолютное значение - так можно понять, насколько сложный ивент. Плюс можно добавить ачивку за прохождение без откатов и смотреть на статистику, сколько таких игроков.

Опасность механики с откатом заключается в том, что игрок может расстроиться из-за ряда неудач и просто покинуть ивент.

Стоит подумать о защите от отката на 3 и более уровня. Можно добавить шуточную ачивку для совсем уж откатных: "Танец маленьких нубят, лвл прямо - пять назад" ("Танец маленьких утят, шаг вперед и два назад"). Смотреть, сколько таких получили, сделать вывод о сложности ивента.

*метрики пользовательской активности за определённый период: за день (daily active users — DAU), за неделю (weekly active users — WAU) или за месяц (monthly active users — MAU).
